```python
# Data manipulation
import pandas as pd
import numpy as np

# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Set configurations
sns.set_style('whitegrid')
```

```python
# Import dataset
args = pd.read_csv('raw/isolates_args.csv', index_col=0)

# Visualize table
args.sample(10)
```

```python
# Build species sub-dataframe
species = args[['Site', 'Isolate', 'Media', 'Species']].drop_duplicates()

# Set figure
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Count species
sns.countplot(x='Species', hue='Media', data=species, palette='summer', alpha=0.7, ax=ax[0])
sns.countplot(x='Species', hue='Site', data=species, palette='summer', alpha=0.7, ax=ax[1])

# Add titles
ax[0].title.set_text('Species by Isolation Media')
ax[1].title.set_text('Species by Sampling Site')

# Add labels
ax[0].set_ylabel('Isolate count')
ax[1].set_ylabel('Isolate count')

# Set italic for species
plt.xticks(fontstyle='italic')
plt.xticks(fontstyle='italic')

# Show figure
plt.show()
```

```python
# Phenotype and genotype
phen = args[args['Type'] == 'Fenotipo'][['Media', 'Name', 'Site']]

# Set figure
fig, ax = plt.subplots(2, 1, figsize=(14, 10))

# Count genotypes and phenotypes
sns.countplot(x='Name', hue='Media', data=phen, palette='summer', alpha=0.7, ax=ax[0])
sns.countplot(x='Name', hue='Site', data=phen, palette='summer', alpha=0.7, ax=ax[1])

# Add labels
ax[0].set_ylabel('Phenotype count')
ax[1].set_ylabel('Phenotype count')

# Limits
ax[0].set_ylim(0, 9)
ax[1].set_ylim(0, 8)

# Add labels
ax[0].set_xlabel('Resistance Phenotype')
ax[1].set_xlabel('Resistance Phenotype')
```
